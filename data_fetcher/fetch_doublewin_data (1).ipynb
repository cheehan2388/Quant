```python
import requests

extra_info_url = "https://api2.bybit.com/s1/byfi/get-products-extra-info"

resp = requests.post(url=extra_info_url, json={"product_type": 10})
all_product_info = resp.json()["result"]["double_win_offers"]
all_product_info
```

```python
product_detail_url = "https://api2.bybit.com/s1/byfi/get-product-detail"

for info in all_product_info:
    resp = requests.post(
        url=product_detail_url,
        json={"product_id": info["product_id"], "product_type": 10},
    )
    print(resp.json()["result"]["double_win_product_detail"])
    info_det = resp.json()["result"]["double_win_product_detail"]["product_name"]
    info["product_name"] = info_det

all_product_info
```
