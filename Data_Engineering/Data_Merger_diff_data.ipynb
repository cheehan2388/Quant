只要左邊的某些數據

```python

# 合併（根據 datetime 欄位，inner join 避免超過的被保留）
merged_df = pd.merge(df1, df2[['datetime', 'Number of trades','Quote asset volume']], on='datetime', how='inner')

# 儲存新的結果
merged_df.to_csv('BTC_merged_V.csv', index=False)
```

我要指定時間的數據

```python
time_start = "2024/5/12"
time_end   = "2025/7/18"
df1_region   = df1[(df1['datetime']>time_start)&(df1['datetime']<time_end)][['datetime','pre_Close']]
df2_region = df2[(df1['datetime']>time_start)&(df2['datetime']<time_end)][['datetime','Close','Volume']]
```

```python
df_combine = pd.merge(df1_region,df2_region,how = 'inner' , on ='datetime')
```
