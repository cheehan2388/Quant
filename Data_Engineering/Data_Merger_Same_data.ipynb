```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('../../CryptoQuant_BTC_Hour_Binance_BTCUSDT_OpenInterest.csv')
df2 = pd.read_csv('../../CG_BTC_H_All_openInterest.csv')

out = 'Binance_Merge_Open_Interest.csv'
 
df1
#df2
```

```python
#merge same data with diff date
df1_unique = df1[~df1['datetime'].isin(df2['datetime'])]


df1_unique
```

```python
combined_df = pd.concat([df1_unique, df2], ignore_index=True)


combined_df = combined_df.sort_values(by='datetime').reset_index(drop=True)

print(combined_df)

combined_df.to_csv(out)
```
